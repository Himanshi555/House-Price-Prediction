# First spilit the data into train and test

```python
from sklearn.model_selection import train_test_split as tts
train, test = tts(data, test_size = 0.2)
```

```python
train_x = train.loc[:, train.columns != "price"]
test_x = test.loc[:, test.columns != "price"]
train_y = train["price"]
test_y = test["price"]
```

# Linear Regression 

In easy words a model in statistics which helps us predicts the future based upon past relationship of variables. So when you see your scatter plot being having data points placed linearly you know regression can help you!

The variable we are predicting is called the criterion variable and is referred to as Y. The variable we are basing our predictions on is called the predictor variable and is referred to as X. When there is only one predictor variable, the prediction method is called
#### Simple Regression. and if multiple predictor variable are present then multiple regression.


# Using Linear Regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(train_x, train_y)
```

# To predict the result

```python
test_pred = model.predict(test_x)
```

```python
result = pd.DataFrame({'Actual': test_y, 'Predicted': test_pred})
print(result)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
print(mean_squared_error(test_y, test_pred))
print(mean_absolute_error(test_y, test_pred))
```

```python
from sklearn.metrics import r2_score
print(r2_score(test_y, test_pred)*100)
```

##### remember that the machine is the student and train data is the syllabus and test data is the exam. we see how much the machine has scored and if it scores well are model is successful.

Letâ€™s go step by step

1. We import our dependencies , for linear regression we use sklearn (built in python library) and import linear regression from it.

3. Now we know that prices are to be predicted , hence we set labels (output) as price columns and we also convert dates to 1â€™s and 0â€™s so that it doesnâ€™t influence our data much . We use 0 for houses which are new that is built after 2014.
4. We again import another dependency to split our data into train and test.

So, We use a different method , which is very important for weak prediction models such as this.

This might seem to be a bit advanced but if understood is a really brilliant tool to enable better predictions.

For building a prediction model , many experts use gradient boosting regression , so what is gradient boosting ? It is a machine learning technique for regression and classification problems, which produces a prediction model in the form of an ensemble of weak prediction models, typically decision trees.

Now to make it easy , remember how we mapped machine as a student , train data as the syllabus and test data as the exam. letâ€™s try to understand gradient boosting method using the same. So , letâ€™s analyse why our student (machine) didnâ€™t get above 85% ? there could be many reasons few such reasons could be :-

Hence for all this problems there is one fix , Gradient descent boosting.


![title](g1.jpeg)

                                           Gradient boosting graph for understanding

```python
from sklearn import ensemble
clf = ensemble.GradientBoostingRegressor(n_estimators = 400, 
      max_depth = 5, min_samples_split = 2,learning_rate = 0.1, loss = 'ls')
```

```python
clf.fit(train_x,train_y)
```

```python
clf.score(test_x,test_y)
```

1. We first import the library from sklearn ( trust me , it is the best library for all statistical related models)
2. We create a variable where we define our gradient boosting regressor and set parameters to it , here
   n_estimatorâ€Šâ€”â€ŠThe number of boosting stages to perform. We should not set it too high which would overfit our model.

   max_depthâ€Šâ€”â€ŠThe depth of the tree node.

   learning_rateâ€Šâ€”â€ŠRate of learning the data.

   lossâ€Šâ€”â€Šloss function to be optimized. â€˜lsâ€™ refers to least squares regression

   minimum sample splitâ€Šâ€”â€ŠNumber of sample to be split for learning the data

3. We then fit our training data into the gradient boosting model and check for accuracy

4. We got an accuracy of 90% which is amazing!
